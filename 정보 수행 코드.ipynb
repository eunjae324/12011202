```python
import csv
import matplotlib.pyplot as plt

year = list(range(2007,2018))
elementary = []
middle = []
high = []

f= open('사교육비.csv')
data = csv.reader(f)
next(data)
next(data)
data = list(data)
for row in data:
    if row[0]=='사교육비':
        for i in range(1,12):
            elementary.append(float(row[2 + 5*(i-1)]))
        
        for i in range(1,12):
            middle.append(float(row[3 + 5*(i-1)]))
            
        for i in range(1,12):
            high.append(float(row[4 + 5*(i-1)]))
               
plt.plot(year, elementary, 'k', label='elementary school') # black
plt.plot(year, middle, 'b', label='middle school') # blue
plt.plot(year, high, 'r', label='high school') # red
plt.xlabel('Year')
plt.ylabel('Private Educational Expenses')
plt.title('Educational Expenses per Student(Monthly Average)')
plt.legend(loc='upper left')
plt.show()
```
